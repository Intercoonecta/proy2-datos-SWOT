```python
import numpy as np
import matplotlib.pyplot as plt
import cartopy
import cartopy.crs as ccrs
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import xarray as xr
# https://docs.xarray.dev/en/stable/gallery.html
import cf_xarray
import folium
import seawater as sw
from palettable import cmocean
import datetime
import json
from erddapy import ERDDAP
```

# Cargando un unico fichero:

```python
url_oleaje_01 = "https://opendap.jpl.nasa.gov/opendap/SeaSurfaceTopography/jason3/preview/L2/GPS-OGDR/c103/JA3_GPSOPR_2PdS103_001_20181124_205057_20181124_224542.nc?time[0:1:6670],lat[0:1:6670],lon[0:1:6670],swh_ku[0:1:6670]"
```

```python
ds = xr.open_dataset(url_oleaje_01)
ds
```

```python
ds.swh_ku.shape
```

```python
plt.figure(figsize=(14,6)) #(17,5)

ax1 = plt.subplot(111, projection=ccrs.PlateCarree())
ds.plot.scatter(x="lon", y="lat", hue="swh_ku", vmax=10, ax=ax1, transform=ccrs.PlateCarree())
# air.plot.contourf(ax=ax)
ax1.coastlines()
gl = ax1.gridlines(crs=ccrs.PlateCarree(), draw_labels=True)
gl.top_labels = False
gl.right_labels = False
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
```

# Cargando varios ficheros:

```python
url_raiz_oleaje_c103 = "https://opendap.jpl.nasa.gov/opendap/SeaSurfaceTopography/jason3/preview/L2/GPS-OGDR/c103/"
```

```python
url_oleaje_01 = "*.nc"
#?time[0:1:6670],lat[0:1:6670],lon[0:1:6670],swh_ku[0:1:6670]
```

```python
url_oleaje_01 = url_raiz_oleaje_c103 + "JA3_GPSOPR_2PdS103_001_20181124_205057_20181124_224542.nc?time[0:1:6670],lat[0:1:6670],lon[0:1:6670],swh_ku[0:1:6670]"
url_oleaje_02 = url_raiz_oleaje_c103 + "JA3_GPSOPR_2PdS103_003_20181124_224541_20181125_004158.nc?time[0:1:6670],lat[0:1:6670],lon[0:1:6670],swh_ku[0:1:6670]"
```

```python
urls_oleaje_c103 = [url_oleaje_01, url_oleaje_02]
```

```python
urls_oleaje_c103
```

```python
ds = xr.open_mfdataset(urls_oleaje_c103)
ds
```

```python
plt.figure(figsize=(14,6)) #(17,5)

ax1 = plt.subplot(111, projection=ccrs.PlateCarree())
ds.plot.scatter(x="lon", y="lat", hue="swh_ku", vmax=10, ax=ax1, transform=ccrs.PlateCarree())
# air.plot.contourf(ax=ax)
ax1.coastlines()
gl = ax1.gridlines(crs=ccrs.PlateCarree(), draw_labels=True)
gl.top_labels = False
gl.right_labels = False
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
```
